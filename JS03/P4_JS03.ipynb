## Langkah 1 - Load Image

```python
from PIL import Image

# Load gambar (pastikan path gambar benar, atau download Lenna.png)
# Untuk contoh, kita asumsikan gambar ada di folder img/
img = Image.open('../Dataset/lena.jpg')
img.show()  # tampilkan gambar
display(img)  # metode alternatif tampilkan gambar
```

## Langkah 2 - Ekstrak Fitur

Ekstrak setiap channel red, green, blue dan hitung histogramnya.

```python
# Ekstrak setiap channel red, green, blue
r, g, b = img.split()

# Cek panjang ukuran channel red
print(len(r.histogram()))

# Cetak fitur histogram pada channel red
print(r.histogram())
```
